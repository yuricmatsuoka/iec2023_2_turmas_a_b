```python
from bing_image_downloader import downloader
import cv2
from google.colab.patches import cv2_imshow
```

```python
def download_images(query, limit=1):
    downloader.download(query, limit=limit, output_dir='images', adult_filter_off=True)

def main():
    objeto = input("Digite o nome do objeto para criar a imagem: ")

    num_images = int(input("Digite o nmero de imagens a serem baixadas: ") or 1)

    download_images(objeto, limit=num_images)

    img_paths = [f'images/{objeto}/Image_{i}.jpg' for i in range(1, num_images + 1)]

    if not img_paths:
        print(f"No foi possvel encontrar imagens para '{objeto}'.")
        return

    print("\nImagens do Objeto:")
    for img_path in img_paths:
        print(img_path)
        img = cv2.imread(img_path)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        cv2_imshow(img_rgb)

if __name__ == "__main__":
    main()
```
